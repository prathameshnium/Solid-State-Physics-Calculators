```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import os
import re
import tkinter as tk
from tkinter import filedialog
from matplotlib import pyplot as plt
```

```python
#frequency
w=np.arange(5,100,0.5)
```

```python
N=10**5
e=1*6
w0=50
V=0.000005
E0=8.854
M=10
n=1
alpha=10**-2
#Er=(N*(e**2)*(w0**2-w**2))/(V*E0*M*((w**2-w**2)**2)+n**2*w**2)
a=(w0**2-w**2)
b=(N*e**2)/(V*E0*M)
Eopt=(N*alpha/V*E0)+1
#Eopt=9+80513036.90763819



Er=(b*a/(a**2+(n*w)**2))+Eopt


Ei=(b*(n*w))/(a**2+(n*w)**2)

```

```python

#Sparkline
#from ipywidgets import interact
#%matplotlib notebook
plt.plot(w,Er,marker=".",)
plt.xlabel('w') 
plt.ylabel('Er') 
plt.tick_params(direction="in",top="on",right="on")

plt.show()
```

```python

plt.plot(w,Ei,marker=".",)
plt.xlabel('w') 
plt.ylabel('Ei') 
plt.tick_params(direction="in",top="on",right="on")
plt.show()
```
